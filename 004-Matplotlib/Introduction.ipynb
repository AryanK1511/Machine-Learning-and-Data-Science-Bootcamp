# Introduction to Matplotlib
Matplotlib is a python library that allows us to turn our data into visualizations. It is built on NumPy arrays just like Pandas. It also integrates directly with Pandas.<br />
**Workflow:**
1. Create data
2. Create plot
3. Plot data and axes on plot
4. Customize plot
5. Save / share plot

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(); # You can add ';' or write plt.show() to only show the plot and not the other output along with it
```

```python
plt.plot([1, 2, 3, 4]);
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Adds axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # Creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # Add some data
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y)
```

### Matplotlib Example Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [23, 54, 78, 98]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # (width, height)

# 3. Plot the data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis");
fig.savefig("assets/sample-plt.png")
```

## Making Figures with NumPy arrays

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # Will return 100 evenly spaces intervals between 0 and 10
x
```

```python
# Plot the data
fig, ax = plt.subplots() # Creates a line plot by default
ax.plot(x, x**2);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
x = np.random.randn(1000);
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Two options to create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
```

```python
ax1.plot(x, x / 2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
fig
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x / 2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```
