```python
# Column from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales["Seats"] = seats_column # Would appear on the right side of the data frame by default
```

```python
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 3.0, 4.2, 9.4, 3.2, 8.2] # Has to be the same length as the data unlike a series
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# Create a column from another column
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
car_sales["Number of wheels"] = 4
```

```python
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

```python
# Removing columns
car_sales = car_sales.drop("Total fuel used", axis=1)
car_sales
```

```python
car_sales = car_sales.drop("Passed road safety", axis = 1)
car_sales
```

```python
# Shuffle your dataframe using sample
car_sales = pd.read_csv("car-sales.csv").sample(frac=1)
```

```python
car_sales
```

```python
car_sales.reset_index(drop = True, inplace = True) # If drop = False, the function would add an extra column with the original indices
```

```python
car_sales
```

```python
# YOu can apply any function to a certain column
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x : x / 1.6)
car_sales
```
