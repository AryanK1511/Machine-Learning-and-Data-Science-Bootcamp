## Machine Learning and Data Science Framework


### Types of Machine Learning
1. __Supervised Learning__:
   - The data we recieve already has categories.
   - Supervised learning is called supervised learning because you have data and labels.
   - A machine learning algorithm tries to use the data to predict a label. If it guesses the label wrong, the algorithm corrects itself and tries again.
   - __Classification__: Classification involves predicting if something is one thing or another. This could be binary or multi-class.
   - __Regression__: Regression problems involve trying to predict a number. You might hear it referred to as a continuous number as well, which just means a number which can go up or down.
3. __Unsupervised Learning__:
   - The data we recieve doesn't have categories. Thre are no labels.
   - Ex) Clustering, Association Rule Learning
   - __Clustering__: Putting groups of similar examples together.
5. __Reinforcement Learning__: It is all about teaching machines through trial and error.
6. __Transfer Learning__: Transfer learning leverages what one machine learning model has learned in another machine learning.

### Types of Data
1. __Structured Data:__ Ex) Excel / CSV Files
2. __Unstructured Data:__ Ex) Pictures / Audio Files
3. __Static Data:__ Doesn't change over time.
4. __Streaming Data:__ Data changes over time.

### Types of Evaluation Metrics
1. __Classification:__ Accuracy, Precision, Recall
2. __Regression:__ Mean absolute error (MAE), Mean squared error (MSE), Root mean squared error (RMSE)
3. __Recommendation:__ Precision at K

### Features in Data
It means what we already know about the data that we have or are using.
- __Feature Variables__: We use these input variables to make predictions or classifications.
- __Target Variable__: It is the output or variable that we are trying to predict.
- __Numerical Features__: Features / variables that are numbers.
- __Categorical Features__: Means one thing or another.
- __Derived Features__: Create a feature using the other features. This process is called feature engineering.
- __Feature Coverage__: Ideally, every sample has some features. There should be more than 10% at least.

### Modelling
1. Choosing and training the model
2. Tuning a model
3. Model comparison
   
##### Splitting Data
- 3 Sets - Training, Validation and Test
- Generalization - The ability of a ML model to perform well on data it hasn't seen before.
- __Tip:__ Traning Split (70% - 80%), Validation Split (10% - 15%), Test Split (10% - 15%)

##### Picking the model
- Your main goal will be knowing what kind of machine learning algorithm to use with what kind of problem this is because some algorithms work better than others on different types of data.
- If you're working with structured data, decision trees such as random forests and gradient boosting algorithms like Cat Boost and XG Boost tend to work best and if you're working with unstructured data, deep learning neural networks and transfer learning tend to work best.
- The goal while training is to minimize the times between experiments.

##### Tuning
- This can be done using the validation set of your data but you can also use the training set for this.
- The goal is to improve your model's performance.

##### Comparison
- Carried out on the test set.
- __Underfitting__: Training data set performance is higher than the test set performance.
- __Overfitting__: Test set performance is higher than the training set performance.
- We wanna have a balance if both for it to be a good model.
- Overfitting can happen due to your test data being leaked into your training data.
- Data mismatch happens when your training data is different from the test data.
- __Underfitting Fix__: Try a more advanced model, Increase model hyperparameters, Reduce amount of features, Train longer.
- __Overfitting Fix__: Collect more data, Try a less advanced model
